```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/heart_failure_clinical_records.csv")
```

```python
plt.scatter(df['age'], df['platelets'], marker='.', edgecolor='red')
plt.ylabel('Platelets')
plt.xlabel('Age')
plt.title('Age vs Platelets')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.boxplot(data=df, orient="v", palette="Pastel1")
plt.title('Boxplot of Platelets')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
